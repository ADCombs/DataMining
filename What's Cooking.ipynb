### Cuisine Analysis

<hr>


```python
cuisines_distinct = df.cuisine.unique() # Returns array of unique values in 'cuisine' column
cuisines_distinct.sort()

print("Total number of distinct cuisines: %d\n" % len(cuisines_distinct))
for cuisine in cuisines_distinct:
    print(cuisine)
```

```python
cuisines_all = df['cuisine']
cuisines_all = list(cuisines_all)

# Test print cuisines_all just to make sure we're counting duplicates

for x in range(0,50):
    print(cuisines_all[x])
```

```python
hist_results = plt.hist(cuisines_all, 20)
plt.xticks(rotation=80)
plt.title('Cuisine Frequency')
plt.xlabel('Cuisine')
plt.show()
```
